## Part f) Lipkim Hamiltonian

### Introduction

In this part we will study the Lipkin Hamiltonian for the case of
$3\times 3$ matrix case with total spin $J=1$


The Hamiltonian in second quantization formulation is given by

$$
H= H_0 + H_1 +H_2,
$$

with

$$
H_0 = \frac{1}{2}\epsilon \sum_{p\sigma} \sigma \alpha_{p\sigma}^\dagger \alpha_{p\sigma},
$$

and

$$
H_1 = \frac{1}{2}V\sum_{p,p',\sigma} \alpha_{p\sigma}^\dagger \alpha_{p'-\sigma}^\dagger \alpha_{p'\sigma} \alpha_{p-\sigma},
$$

and

$$
H_2 = \frac{1}{2}W\sum_{p,p',\sigma} \alpha_{p\sigma}^\dagger \alpha_{p'-\sigma}^\dagger \alpha_{p'\sigma} \alpha_{p-\sigma}
$$

We can rewrite the above Hamiltonian in terms the so-called quasispin operators
leading to

$$
H_0 = \epsilon J_z
$$

and

$$
H_1 = -(\frac{V}{2}J_+^2 + \frac{V}{2}J_-^2),
$$

We start here with simple case, namly $J=1$ case and we set $W=0$.
First we will to show

$$
\begin{equation}
H_{J=1}= \begin{bmatrix}
                -\epsilon & 0 & -V \\
                0  & 0 & 0 \\
                -V & 0 & \epsilon
                \end{bmatrix}.
\end{equation}
$$

Derive the Hamiltonian in the $J=1$ case.

For a system with totall spin of $J=1$ the system have three eigenvalues for the $z$ component($J_z$) of the total spin operator. The eigenvalues are $J_z=-1,0,1$. we can thus define the eigen states for these eigenvalues

$$
\ket{J,J_z} = \ket{1,1}, \ket{1,0}, \ket{1,-1}.
$$

The operator $J_z$ is diagonal in the basis of the eigenstates of $J_z$

$$
J_z\ket{J,J_z} = J_z\ket{J,J_z}
$$

For our three states we have

$$

J_z\ket{1,-1} = -1 \ket{1,-1}\\
J_z\ket{1,0} = 0 \ket{1,0}\\
J_z\ket{1,1} = 1\ket{1,1}


$$

In the matrix form with the basis $\{\ket{1,1},\ket{1,0},\ket{1,-1}\}$ we have

$$
J_z = \begin{bmatrix}
                -1 & 0 & 0 \\
                0 & 0 & 0 \\
                0 & 0 & 1
                \end{bmatrix}.
$$

Next we will construct the full matrix representation of the raising and lowering operators $J_+$ and $J_-$.
by looking at what happens when we apply the raising and lowering operators to each of the eigenstates of $J_z$.

The action of the raising operator $J_+$ is given by

$$
J_+ \vert J,J_z\rangle = \sqrt{J(J+1) - J_z(J_z + 1)} \vert J,J_z + 1\rangle,
$$

And the action of the lowering operator $J_-$ is given by

$$
J_- \vert J,J_z\rangle = \sqrt{J(J+1) - J_z(J_z - 1)} \vert J,J_z - 1\rangle.
$$

The above relations are retrived from week7 lecture notes equation (19) and (20).
https://github.com/CompPhysics/QuantumComputingMachineLearning/blob/gh-pages/doc/pub/week7/ipynb/week7.ipynb

Let us start with the raising operator $J_+$.

$$
J_+ \ket{1,-1} = \sqrt{1(1+1) - (-1)(-1 + 1)} \ket{1,0} = \sqrt{2} \ket{1,0}
$$

$$
J_+ \ket{1,0} = \sqrt{1(1+1) - 0(0 + 1)} \ket{1,1} = \sqrt{2} \ket{1,1}
$$

$$
J_+ \ket{1,1} = \sqrt{1(1+1) - 1(1 + 1)} \ket{1,2} = 0 \ket{1,2}
$$

Observing the above we see that the first basis state is mapped to the second basis state with a factor of $\sqrt{2}$, the second basis state is mapped to the third basis state with a factor of $\sqrt{2}$ and the third basis state has no mapping to any state.
Thus we can write the matrix representation of the raising operator $J_+$ in the basis $\{\ket{1,-1},\ket{1,0},\ket{1,1}\}$ as

$$
J_+ = \begin{bmatrix}
                0 & 0 & 0 \\
                \sqrt{2} & 0 & 0 \\
                0 & \sqrt{2} & 0
                \end{bmatrix}.
$$


Now we do the same for the lowering operator $J_-$.

$$
J_- \ket{1,1} = \sqrt{1(1+1) - 1(1 - 1)} \ket{1,0} = \sqrt{2} \ket{1,0}
$$

$$
J_- \ket{1,0} = \sqrt{1(1+1) - 0(0 - 1)} \ket{1,-1} = \sqrt{2} \ket{1,-1}
$$

$$
J_- \ket{1,-1} = \sqrt{1(1+1) - (-1)(-1 - 1)} \ket{1,-2} = 0 \ket{1,-2}
$$

Observing the above we see that the third basis state is mapped to the second basis state with a factor of $\sqrt{2}$, the second basis state is mapped to the first basis state with a factor of $\sqrt{2}$ and the first basis state has no mapping to any state.


Thus we have the matrix representation of the lowering operator $J_-$ in the basis $\{\ket{1,-1},\ket{1,0},\ket{1,1}\}$ as

$$
J_- = \begin{bmatrix}
                0 & \sqrt{2} & 0 \\
                0 & 0 & \sqrt{2} \\
                0 & 0 & 0
                \end{bmatrix}.
$$

Next we have to compute $J_+^2$ and $J_-^2$.

$$
J_+^2 = J_+ J_+ = \begin{bmatrix}
                0 & 0 & 0 \\
                \sqrt{2} & 0 & 0 \\
                0 & \sqrt{2} & 0
                \end{bmatrix}
                \begin{bmatrix}
                0 & 0 & 0 \\
                \sqrt{2} & 0 & 0 \\
                0 & \sqrt{2} & 0
                \end{bmatrix}
                =
                \begin{bmatrix}
                0 & 0 & 0 \\
                0 & 0 & 0 \\
                2 & 0 & 0
                \end{bmatrix}
$$

$$
J_-^2 = J_- J_- = \begin{bmatrix}
                0 & \sqrt{2} & 0 \\
                0 & 0 & \sqrt{2} \\
                0 & 0 & 0
                \end{bmatrix}
                \begin{bmatrix}
                0 & \sqrt{2} & 0 \\
                0 & 0 & \sqrt{2} \\
                0 & 0 & 0
                \end{bmatrix}
                =
                \begin{bmatrix}
                0 & 0 & 2 \\
                0 & 0 & 0 \\
                0 & 0 & 0
                \end{bmatrix}
$$

Our Hamiltonian equation (1) is given by

$$
H_{J=1} = \epsilon J_z + c
$$

Here we dont add the $W$ term since, for now we set it to zero.
We can now plug in the matrices we have computed above into the Hamiltonian.

$$
H_{J=1} = \epsilon \begin{bmatrix}
                -1 & 0 & 0 \\
                0 & 0 & 0 \\
                0 & 0 & 1
                \end{bmatrix}
                +
                \frac{V}{2}\begin{bmatrix}
                0 & 0 & 0 \\
                0 & 0 & 0 \\
                2 & 0 & 0
                \end{bmatrix}
                +
                \frac{V}{2}\begin{bmatrix}
                0 & 0 & 2 \\
                0 & 0 & 0 \\
                0 & 0 & 0
                \end{bmatrix}
$$

$$
=
\begin{bmatrix}
                -\epsilon & 0 & V \\
                0  & 0 & 0 \\
                V & 0 & \epsilon
                \end{bmatrix}.
$$

This is the matrix representation of the Lipkin Hamiltonian for the case of $J=1$.
We notice that there is missing a minus sign in the elements of V, I think this is becasue
there should be a minus sign in the $H_1$ term in the Hamiltonian.

$$
\begin{equation}
H_1 = - \frac{1}{2} V \left( J_+^2 + J_-^2 \right).
\end{equation}
$$

Thus
we have our Hamiltonian for the case of $J=1$ as

$$
H_{J=1} = \begin{bmatrix}
                -\epsilon & 0 & -V \\
                0  & 0 & 0 \\
                -V & 0 & \epsilon
                \end{bmatrix}.
$$


Now it is easy construct the Hamiltonian when $W\neq 0$ For that we just need construct the $H_2$ using Equation (15) in https://journals-aps-org.ezproxy.uio.no/prc/abstract/10.1103/PhysRevC.106.024319


$$
J_+J_- = \begin{bmatrix}
                0 & 0 & 0 \\
                0 & 2 & 0 \\
                0 & 0 & 2
                \end{bmatrix}
$$
and
$$
J_-J_+ = \begin{bmatrix}
                2 & 0 & 0 \\
                0 & 2 & 0 \\
                0 & 0 & 0
                \end{bmatrix}
$$

Thus we have
$$
H_{J=1} = \begin{bmatrix}
                -\epsilon & 0 & -V \\
                0  & 0 & 0 \\
                -V & 0 & \epsilon
                \end{bmatrix}
                -
                \frac{W}{2}\begin{bmatrix}
                2 & 0 & 0 \\
                0 & 2 & 0 \\
                0 & 0 & 0
                \end{bmatrix}
                -
                \frac{W}{2}\begin{bmatrix}
                0 & 0 & 0 \\
                0 & 2 & 0 \\
                0 & 0 & 2
                \end{bmatrix}
$$

$$
=
\begin{bmatrix}
                -\epsilon - W & 0 & -V \\
                0  & -2W & 0 \\
                -V & 0 & \epsilon - W
                \end{bmatrix}.
$$


```python
from quantum_computing.p1.hamiltonian.hamiltonian import LipkinHamiltonian
from quantum_computing.p1.quantum.utils import plot_eigenvalues
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = [7, 3]
plt.rcParams['font.size'] = 12
```

## Construct the Hamiltonian for the case of $J=1$ with $W= 0$

```python
epsilon_j1 = 1
V = 0.5
W = 0

lipkin = LipkinHamiltonian()

H_j1 = lipkin.create_j1_hamiltonian(epsilon_j1,V,W)
print(H_j1)
```

### Solve the eigenvalue problem

```python

v_values = np.linspace(0, 2, 100)

eigenvalues, eigenvectors = lipkin.solve_eigen_vs_V(1,epsilon_j1,v_values,W)

plot_eigenvalues(v_values,eigenvalues,1,"Eigenvalue Spectrum of Lipkin model (J=1)")
plt.show()
```

Given that the $W$ parameter is set to zero, we get that for the
when we start with $V=0$ the eigenvalues are only $J_z$ projections.
The eigenvalues are $-1,0,1$. But as we increase the $V$ parameter
the eigenvalue for $J_z=1$ and $J_z=-1$ increase their energy gap. while the
$J_z=0$ eigenvalue is not affected by the $V$ parameter.
This can clearly be interpreted as the second eigenstate from the Hamiltonian matrix column 2 is always being 
zero energy, given that $W=0$.


```python
# For J=1 and W=0, the ground state
ground_states = [state[:,0] for state in eigenvectors]
ground_state_probabilities = np.array([np.abs(ground_state)**2 for ground_state in ground_states])


plot_eigenvalues(v_values,ground_state_probabilities,1,"Ground State Probabilities for (J=1,W=0) Lipkin Model")
plt.show()
```

Observing the probability amplitude of the ground eigenstate we see that as we increase the $V$ parameter There appears more and more mixing of the $J_z=-1$ and $J_z=1$ states. while the $J_z=0$ state is not affected by the $V$ parameter. There is no clear phase transition.
We see that for interaction strength $V=0$ we see that the ground state is populated only by the first eigenstate $J_z=-1$

### Solve the eigenvalue problem for $W\neq 0$

```python
epsilon_j1 = 1
V = 0.5
W = 1.5

lipkin = LipkinHamiltonian()
H_j1 = lipkin.create_j1_hamiltonian(epsilon_j1,V,W)
print(H_j1)
```

```python
v_values = np.linspace(0, 2, 100)
eigenvalues, eigenvectors = lipkin.solve_eigen_vs_V(1,epsilon_j1,v_values,W)
plot_eigenvalues(v_values,eigenvalues,1,"EigenvaluesEigenvalues of the Lipkin Hamiltonian (J=1,W=1.5)")

plt.show()
```

When we set $W=1.5$ We se that there is a degenerate solution 
at around $V=1.125$ Similar to what we found for the Hamiltonian in part d and part e. For $W=1.5$ The degenerate appears for eigenstate $J_z=0$ and $J_z=-1$. However if we set $W=-1.5$ we see that the degenerate solution appears for $J_z=0$ and $J_z=1$. The reader should change the parameter and explore.

```python
# For J=1 and W=1.5, the ground state
ground_states = [state[:,0] for state in eigenvectors]
ground_state_probabilities = np.array([np.abs(ground_state)**2 for ground_state in ground_states])

plot_eigenvalues(v_values,ground_state_probabilities,1,"Ground State Probabilities for (J=1,W=1.5) Lipkin Model")
plt.show()
```

Analyzing the probability amplitudes for the ground state we see that 
up until V = 1.125 the ground state 100% consists of the $J_z=0$ state.
And then the system undergoes a phase transition then the states consists of a combination of $J_z=-1$ and $J_z=1$ states.
An interersting notice is that the ground state is not populated by the $J_z=-1$ state with $V=0$ anymore like in the case for $W=0$ above. 
The action of the $W$ parameter changes the behavior of the system and at the same time adds phase transition points in the system for the values of $V$ 

### Solve the eigenvalue problem for $J=2$ and $W=0$

```python
epsilon_j2 = 1
V = 0
W = 0

lipkin = LipkinHamiltonian()

H_j2 = lipkin.create_j2_hamiltonian(epsilon_j2,V,W)
print("Hamiltonian")
print(H_j2)
print()
eigenvalues, eigenvectors = lipkin.solve_eigenvalue_problem(H_j2)
print("Eigen values: ",eigenvalues)
```

We have above constructed the Hamiltonian for the case of $J=2$ and $W=0$. And solved the eigenvalue problem for a choosen value of
$V=0$. This then directly gives us the eigenvalues for basis states of $J_z=-2,-1,0,1,2$. being the diagonal elements of the Hamiltonian matrix. 

Next we will solve for eigenvalues and eigenstates for as a function of $V=[0,2]$

```python

v_values = np.linspace(0, 2, 100)
eigenvalues, eigenvectors = lipkin.solve_eigen_vs_V(2,epsilon_j2,v_values,W)

plot_eigenvalues(v_values,eigenvalues,2,"EigenvaluesEigenvalues of the Lipkin Hamiltonian (J=2,W=1.5)")
plt.show()
```

With $J=2$ and $W=0$ We get similar reults as the $J=1$ case, but this time we have five energi states where state $J_z=-2$ and $J_z=-1$ 
decrease with increasing $V$ parameter. The $J_z=0$ state is not affected by the $V$ parameter. And 
the $J_z=1$ and $J_z=2$ states increase with increasing $V$ parameter. 

Next we study the groundstate of the system for the $V$ values

```python
# For J=2, W=0 case
v_values = np.linspace(0, 2, 100) 

ground_states = [state[:,0] for state in eigenvectors]
ground_state_probabilities = np.array([np.abs(ground_state)**2 for ground_state in ground_states])

plot_eigenvalues(v_values,ground_state_probabilities,2,"Ground State Probabilities for (J=2,W=1.5) Lipkin Model")
plt.show()
```

Interpreting the above figure yeilds same notion that for W=0 as we increase the $V$ parameter there appears more and more mixing of the the states, however not any clear phase transition.

### Solve the eigenvalue problem for $J=2$ and $W\neq 0$

```python
W = 1.5
v_values = np.linspace(0, 2, 100)

eigenvalues, eigenvectors = lipkin.solve_eigen_vs_V(2,epsilon_j2,v_values,W)

plot_eigenvalues(v_values,eigenvalues,2,"Eigenvalues of the Lipkin Hamiltonian (J=2,W=1.5)")
plt.show()
```

When we set $W=1.5$ We see that there appears degenerate solutions, the choise of $W$ parameter we see states $J_z=-2$ and $J_z=-1$ are degenerate.
The $J_z=0$ and $J_z=1$ states are also degenerate. Setting the $W$ parameter to $-1.5$ we see the the higher states undergoes a degenerate solution.

### Studying the probabilities if the ground state

```python
# For J=2 case and W=1.5
ground_states = [state[:,0] for state in eigenvectors]
ground_state_probabilities = np.array([np.abs(ground_state)**2 for ground_state in ground_states])

plot_eigenvalues(v_values,ground_state_probabilities,2,"Ground State Probabilities for (J=2,W=1.5) Lipkin Model")
plt.show()
```

For our choice of parameters we clearly see that there are two phase transitions happening.
at around $V=1.125$ and $V=1.45$.

the ground state is mostly taken by the $J_z=0$ state, and the amplitude decreases with increasing $V$ at the first phase transition point a complex mixing appears and after the second phase transition it appears that the system transistions back to the same state as before the first phase transition.
